```python
# import packages
import pandas as pd
import numpy as np

# load in data
waterCSV = pd.read_csv("..\data\waterSource.csv")
disCSV = pd.read_csv("..\data\disclosureList.csv")
```

```python
# pivot waterCSV on Description column and replace NA values with 0

wp = waterCSV.pivot(columns = "Description", index = ["DisclosureId", "APINumber"], values = "Percent").fillna(value = 0).reset_index()

# aggregate and rename columns 
wp["ground"] = wp[["Groundwater, < 1000TDS", "Groundwater, > 1000TDS"]].sum(axis = 1)
wp["surface"] = wp[["Surface Water, < 1000TDS", "Surface Water, > 1000TDS"]].sum(axis = 1)
wp["other"] = wp[["Other, < 1000TDS", "Other, > 1000TDS"]].sum(axis = 1)
wp["produced"] = wp["Produced Water"]

# select only columns relevant to analysis
wp2 = wp[["DisclosureId", "ground", "surface", "produced", "other"]]
```

```python
# join disclosures list and water source information into single dataframe
cjn = pd.merge(disCSV, wp2, how = "inner", on = "DisclosureId")

# select only columns relevant to analysis
cjn = cjn[['DisclosureId', 'APINumber', 'StateName','CountyName', 
           'OperatorName', 'WellName', 'Latitude', 'Longitude',
           'TotalBaseWaterVolume', 'TotalBaseNonWaterVolume',
           'FederalWell', 'IndianWell', 'ground', 'surface',
           'produced', 'other']]

cjn
```
